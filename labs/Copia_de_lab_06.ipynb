1. Gráfique mediante un gráfico tipo scatter cada grupo. A simple vista, ¿ los grupos son muy distintos entre si?.

```python
grupo1=df.loc[df['grupo']=='Grupo_1']
grupo2=df.loc[df['grupo']=='Grupo_2']
grupo3=df.loc[df['grupo']=='Grupo_3']
grupo4=df.loc[df['grupo']=='Grupo_4']

plt.scatter(grupo1['x'], grupo1['y'], c='grey', label='Grupo 1')
plt.xlabel('x1')
plt.ylabel('y1')
plt.title('datos del Grupo 1')
plt.legend()
plt.show()

plt.scatter(grupo2['x'], grupo2['y'], c='grey', label='Grupo 2')
plt.xlabel('x2')
plt.ylabel('y2')
plt.title('datos del Grupo 2')
plt.legend()
plt.show()

plt.scatter(grupo3['x'], grupo3['y'], c='grey', label='Grupo 3')
plt.xlabel('x3')
plt.ylabel('y3')
plt.title('datos del Grupo 3')
plt.legend()
plt.show()

plt.scatter(grupo4['x'], grupo4['y'], c='grey', label='Grupo 4')
plt.xlabel('x4')
plt.ylabel('y4')
plt.title('datos del grupo 4')
plt.legend()
plt.show()
```

A simple vista, se logra observar que los datos de estos grupos son muy diferentes, pues los graficos obtenidos son muy diferentes

3. Realice un ajuste lineal para cada grupo. Además, grafique los resultados de la regresión lineal para cada grupo. Interprete.

```python
from sklearn.linear_model import LinearRegression

modelo1= LinearRegression()
modelo1.fit(grupo1[['x']], grupo1['y'])
mod1_pred=modelo1.predict(grupo1[['x']])

plt.scatter(grupo1['x'], grupo1['y'], c='grey', label='Grupo 1')
plt.plot(grupo1['x'], mod1_pred, c='black', linestyle='-', label='prediction modelo 1')
plt.xlabel('x1')
plt.ylabel('y1')
plt.title('datos del Grupo 1')
plt.legend()
plt.show()
```

Al hacer la regresion lineal de los datos del grupo 1, se logra observar que estos si siguen un ajuste lineal y este modelo es adecuado para este grupo

```python
modelo2= LinearRegression()
modelo2.fit(grupo2[['x']], grupo2['y'])
mod2_pred=modelo2.predict(grupo2[['x']])

plt.scatter(grupo2['x'], grupo2['y'], c='grey', label='Grupo 2')
plt.plot(grupo2['x'], mod2_pred, c='black', linestyle='-', label='prediction modelo 2')
plt.xlabel('x2')
plt.ylabel('y2')
plt.title('datos del Grupo 2')
plt.legend()
plt.show()
```

Al hacer la regresion lineal de los datos del grupo 2, se logra observar que hay una buena cantidad de puntos que no se ajustan a este modelo y no es el mejor modelo para este grupo

```python
modelo3= LinearRegression()
modelo3.fit(grupo3[['x']], grupo3['y'])
mod3_pred=modelo3.predict(grupo3[['x']])

plt.scatter(grupo3['x'], grupo3['y'], c='grey', label='Grupo 3')
plt.plot(grupo3['x'], mod3_pred, c='black', linestyle='-', label='prediction modelo 3')
plt.xlabel('x3')
plt.ylabel('y3')
plt.title('datos del Grupo 3')
plt.legend()
plt.show()
```

Al hacer la regresion lineal de los datos del grupo 3, se logra observar casi todos los puntos siguen un ajuste lineal pero debido a un punto muy alejado de los demas, provoca un desajuste y afecta en la regesion lineal en este grupo

```python
modelo4= LinearRegression()
modelo4.fit(grupo4[['x']], grupo4['y'])
mod4_pred=modelo4.predict(grupo4[['x']])

plt.scatter(grupo4['x'], grupo4['y'], c='grey', label='Grupo 4')
plt.plot(grupo4['x'], mod4_pred, c='black', linestyle='-', label='prediction modelo 4')
plt.xlabel('x4')
plt.ylabel('y4')
plt.title('datos del Grupo 4')
plt.legend()
plt.show()
```

Al hacer la regresion lineal de los datos del grupo 4, se logra observar que un ajuste lineal no es el mejor modelo para este grupo pues tiene muchos datos alejados de la regresion linela obtenida. Por otro lado, se logra observar dos conjuntos de grupos, donde si eliminamos el punto alejado del resto de puntos y cambiamos los valores x por los valores y del otro grupo, un ajuste lineal seria lo adecuado.

4. Calcule los resultados de las métricas para cada grupo. Interprete.

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error
from sklearn. metrics import r2_score

def smape(y_true, y_pred):
  return round(np.mean(np.abs(y_pred-y_true)/((np.abs(y_pred)**2 + np.abs(y_true)**2)/2)), 2)

print("Errores del Grupo 1:")
print("MAE:", round(mean_absolute_error(grupo1['y'], mod1_pred),2))
print("MSE:", round(mean_squared_error(grupo1['y'], mod1_pred),2))
print("MAPE:", round(mean_absolute_percentage_error(grupo1['y'], mod1_pred),2))
print("SMAPE:", smape(grupo1['y'], mod1_pred))
print("R2:", round(r2_score(grupo1['y'], mod1_pred),2))
print("\n")

print("Errores del Grupo 2:")
print("MAE:", round(mean_absolute_error(grupo2['y'], mod2_pred),2))
print("MSE:", round(mean_squared_error(grupo2['y'], mod2_pred),2))
print("MAPE:", round(mean_absolute_percentage_error(grupo2['y'], mod2_pred),2))
print("SMAPE:", smape(grupo2['y'], mod2_pred))
print("R2:", round(r2_score(grupo2['y'], mod2_pred),2))
print("\n")

print("Errores del Grupo 3:")
print("MAE:", round(mean_absolute_error(grupo3['y'], mod3_pred),2))
print("MSE:", round(mean_squared_error(grupo3['y'], mod3_pred),2))
print("MAPE:", round(mean_absolute_percentage_error(grupo3['y'], mod3_pred),2))
print("SMAPE:", smape(grupo3['y'], mod3_pred))
print("R2:", round(r2_score(grupo3['y'], mod3_pred),2))
print("\n")

print("Errores del Grupo 4:")
print("MAE:", round(mean_absolute_error(grupo4['y'], mod4_pred),2))
print("MSE:", round(mean_squared_error(grupo4['y'], mod4_pred),2))
print("MAPE:", round(mean_absolute_percentage_error(grupo4['y'], mod4_pred),2))
print("SMAPE:", smape(grupo4['y'], mod4_pred))
print("R2:", round(r2_score(grupo4['y'], mod4_pred),2))
```

Con los resultados obtenidos, se logra observar que los valores de metrica del grupo 3 son los mas bajos entre los cuatro grupos, lo cual indica que un ajuste lineal es un buen modelo para este grupo. Por otro lado, los valores de metrica del grupo 2 son los mas alto entre los cuatro grupos, lo cual indica que un ajuste lineal no es un modelo para este grupo sin hacer cambios en estos datos. 

5. Es claro que el ajuste lineal para algunos grupos no es el correcto. Existen varias formas de solucionar este problema (eliminar outliers, otros modelos, etc.). Identifique una estrategia para que el modelo de regresión lineal ajuste de mejor manera e implemente otros modelos en los casos que encuentre necesario.

En el grupo 1, se logra apreciar que un ajuste lineal no es modelo completamente adecuado para este grupo, pero posee pocos puntos de este grupo que están muy alejados del ajuste lineal. Si se usara otro modelo, se podria producir un sobreajuste al intentar que este lo mas cerca posible de todos los puntos del grupo y no seria muy bueno prediciendo resultados.

Por otro lado, se logra apreciar que los puntos del grupo 2 tiene comportamiento de una parabola. Entonces haremos un ajuste polinomial de grado 2 para este grupo.

```python
from sklearn.preprocessing import PolynomialFeatures

poli=PolynomialFeatures(degree=2)
nuevo_grupo2=grupo2.sort_values('x')
x_poli=poli.fit_transform(nuevo_grupo2[['x']])
poli_reg= LinearRegression()
poli_reg.fit(x_poli, nuevo_grupo2['y'])
y_pred2=poli_reg.predict(x_poli)


plt.scatter(nuevo_grupo2['x'], nuevo_grupo2['y'], c='grey', label='Grupo 2')
plt.plot(nuevo_grupo2['x'], y_pred2, c='black', label='prediction nuevo modelo 2')
plt.xlabel('x2')
plt.ylabel('y2')
plt.title('datos del Grupo 2')
plt.legend()
plt.show()
```
